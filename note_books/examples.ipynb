```python
import plotly.graph_objects as go

def plot_activity_critical_power(df: pd.DataFrame)->go.Figure:
    """ Plot the critical power of an activity
    df: with df being the critical power dataframe from critical_power(df)['df']
    returns: fig
    """
    fig = go.Figure()
    
    upper_std = df['cp'] + df['std']
    lower_std = df["cp"] - df['std']
    
    fig.add_trace(go.Scatter(x=df['seconds'], y=df['cp'], fill=None, mode='lines', line_color='red', name='Critical Power'))
    fig.add_trace(go.Scatter(x=df['seconds'], y=upper_std, fill='tonexty', mode='lines', line_color='lightblue', name='STD'))
    fig.add_trace(go.Scatter(x=df['seconds'], y=lower_std, fill='tonexty', mode='lines', line_color='lightblue', name='STD'))
    fig.update({
        "layout": {
            "title": "Critical Power",
            "xaxis": {
                "title": "Time (s)"
            },
            "yaxis": {
                "title": "Power (W)"
            }
        }
    })

    return fig
fig = plot_activity_critical_power(cp[:1200])
fig.show()
```
