### Notes
- force(newtons) = mass(kg) * acceleration(m/s^2)
- energy(joules) = force(newtons) * distance(meters) = mass(kg) * acceleration(m/s^2) * distance(meters)
- power(watts) = energy(joules) / time(seconds) = mass(kg) * acceleration(m/s^2) * distance(meters) / time(seconds)
